```python
import pandas as pd
```

```python
#You need to place the dataset Excel file 'Customer Call List.xlsx' in the same folder as this Jupyter Notebook, the 2 files side by side in the same folder
#in order to be read with this simple code line, without having to specify a local path to this file
df=pd.read_excel('Customer Call List.xlsx')
df
```

```python
# Remove duplicate rows from the dataframe to ensure data integrity with drop_duplicates() method
df1=df.drop_duplicates()
df1
```

```python
#Remove the not useful columns in the dataset, like the last column in this case
df2 = df1.drop(columns='Not_Useful_Column')
df2
```

```python
# Remove leading forward slashes from Last_Name column
# Remove ellipsis (...) from beginning and end of Last_Name values
# Remove trailing underscores from Last_Name values
df2['Last_Name']=df2['Last_Name'].str.lstrip('/')
df2['Last_Name']=df2['Last_Name'].str.strip('...')
df2['Last_Name']=df2['Last_Name'].str.rstrip('_')
df2
# Else you can do everything in one step 
# df2['Last_Name']=df2['Last_Name'].str.strip('123._/')
```

```python
# Let's clean the phone numbers column: clean phone numbers by removing special characters (-, /, |)
df2["Phone_Number"]=df2["Phone_Number"].str.replace(r'[^a-zA-Z0-9]','',regex=True)
#Else to remove all non-digit characters from phone numbers you can use 
#df['phone'] = df['phone'].str.replace(r'[-/|]', '', regex=True)
df2["Phone_Number"]=df2["Phone_Number"].apply( lambda x: str(x))
df2["Phone_Number"]=df2["Phone_Number"].apply( lambda x: x[0:3]+'-'+ x[3:6]+'-'+x[6:10])
df2
```

```python
# Then replace NaN and N/a with actual NULL values
df2["Phone_Number"]=df2["Phone_Number"].replace('nan--','')
df2["Phone_Number"]=df2["Phone_Number"].replace('Na--','')
df2
#df2["Phone_Number"].replace('NaN', np.nan)
```

```python
#Split the Address column into 3 columns for each type of information
df2[['Street_Address','State','Zip_Code']]=df2['Address'].str.split(',', n=2, expand=True)
df2
```

```python
#The selected code is replacing values in two columns of the DataFrame `df2`. Replaces 'Yes'/'No' values with 'Y'/'N'
df2['Paying Customer']=df2['Paying Customer'].replace('Yes','Y')
df2['Paying Customer']=df2['Paying Customer'].replace('No','N')
df2['Do_Not_Contact']=df2['Do_Not_Contact'].replace('Yes','Y')
df2['Do_Not_Contact']=df2['Do_Not_Contact'].replace('No','N')
df2
```

```python
# Replace 'N/a' values with empty strings
df2=df2.replace('N/a','')
#df2=df2.replace('NaN','')
# For NaN you need to use another method because it is not a string, it's a NULL value. Fill all NaN/null values with method fillna('') 
df2=df2.fillna('')
df2
```

```python
# Now we have finished the Data Cleaning column by column, let's finish by cleaning all the Dataframe
```

```python
# Do_not_Contact==Y & missing Phone_Number=''
# You don't contact the Customers who have said Do_not_Contact=Y
# So you are going to drop the rows in question by for looping with a criterion
df2['Do_Not_Contact']=df2['Do_Not_Contact'].replace('','N')
df2['Paying Customer']=df2['Paying Customer'].replace('','N')
df2
```
